```python
import matplotlib.pyplot as plt
import statistics
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
savings = [0, 100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots() # returns a tuple that contains a figure and axes objects.
ax.plot(x_axis, y_axis)  # plots the data from the x and y axes.
```

```python
# following does the same thing. This can be changed for axis labels, title, legend.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create a quick plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

# Use the matlab methods
plt.figure(figsize=(w, h))	Change the size of the figure in pixels. Added on the first line of the script.
plt.plot(x, y, label='line')	Add a label that will be added to the legend.
plt.xlim(min, max)	Set the min and max range of the x-axis.
plt.ylim(min, max)	Set the min and max range of the y-axis.
plt.xlabel('x label')	Add a label to the x-axis.
plt.ylabel('y label')	Add a label to the y-axis.
plt.title("Title")	Add a title.
plt.legend()	Add a legend.
plt.grid()	Add a grid to the chart.
plt.savefig("add a path and figure extension")	Save the figure with the given extension. Added at the end of the script.

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston', marker='*', color='blue',linewidth=2)
# Create labels for x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title('PyBer Fare by Month')
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.figure(figsize=(2000,10000)) # does not work
plt.show()
```

# Use the object-oriented methods
fig, ax = plt.subplots(figsize=(w, h))	Change the size of the figure in pixels. Add this in the subplots() function.
ax.plot(x, y, label='line')	Add a label that will be added to the legend.
ax.set_ylim(min, max)	Sets the min and max range of the y-axis.
ax.set_xlim(min, max)	Sets the min and max range of the x-axis.
ax.set_xlabel('x label')	Add a label to the x-axis.
ax.set_ylabel('y label')	Add a label to the y-axis.
ax.set_title("Title")	Add a title.
ax.legend()	Add a legend.
ax.grid()	Add a grid to the chart.
** plt.savefig("add a path and figure extension")	Saves the figure with the given extension. Added at the end of your script.

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_axis, y_axis, label='Boston', marker='d', color='green', linewidth=2)
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.legend()
ax.grid()
```

```python
# create a vertical bar plot
plt.bar(x_axis, y_axis, color='green', label='Boston')
# create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.ylim(0, 45)
plt.legend()

```

```python
# create a horizontal bar plot
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()  # Invert y axis. gca: get current axes.
# create labels for the x and y axes.
plt.ylabel('Date')
plt.xlabel('Fare ($)')
plt.title('PyBer Fare by Month')
#plt.ylim(0, 45)
plt.legend()
```

```python
# using the object-oriented approach
# create the vertical bar plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='black', label='New York')
ax.legend()
```

```python
# create a horizontal bar plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label='Chicago', color='cyan')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.legend()
```

```python
# create a scatter plot using Matlab method
plt.plot(y_axis, x_axis, 'o', c='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])  # list comprehension: new_list = [expression for item in list if conditional]
```

```python
# create a scatter plot using object-oriented method
fig, ax = plt.subplots()
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
ax.scatter(y_axis, x_axis, label='Boston', c='xkcd:sky blue', alpha=0.8, edgecolors='k', linewidth=2, s=[i*5 for i in y_axis])
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

```python
# create a pie chart using matlab method
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        labels=x_axis, 
        explode=explode_values, 
        colors=colors,
        autopct='%.1f%%')
plt.show()
```

```python
# create a pie chart using object-oriented method
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
       labels=x_axis, 
       explode=explode_values, 
       colors=colors,
       autopct='%.1f%%',
       shadow=True,
      startangle=270)
ax.invert_yaxis()
plt.show()
```

```python
# get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# line plot with error bars using matlab method
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
```

```python
# line plot with error bars using object-oriented method
fig, ax = plt.subplots()
ax.errorbar(
    x_axis,
    y_axis,
    yerr=stdev,
    capsize=5
)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(
    x_axis,
    y_axis,
    yerr=stdev,
    capsize=3
)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0)) 
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
ax.xaxis.set_ticks(np.arange(0,51,step=5.0)) # same as the above line
ax.xaxis.set_minor_locator(MultipleLocator(1)) 
ax.invert_yaxis()
plt.show()
```
